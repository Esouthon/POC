```python
import requests
import csv
import datetime
import sys
# Remplacez 'chemin/vers/votre/projet/module' par le chemin absolu réel vers le dossier "module" via la commande pwd
module_path = '/Users/eliotsouthon/Desktop/POC/POC/Code/Modules'  
sys.path.append(module_path)
import ma_bibliotheque 

def get_weather_forecast_3h(lat, lon, api_key):
    url = f"https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={api_key}&units=metric"
    
    # Requête vers l'API
    response = requests.get(url)
    
    # Vérification du succès de la requête
    if response.status_code == 200:
        data = response.json()
        
        # Extraire les prévisions toutes les 3 heures pour les 5 jours
        forecast_data = []
        for forecast in data['list']:
            forecast_info = {
                "date": datetime.datetime.fromtimestamp(forecast['dt']).strftime('%Y-%m-%d %H:%M:%S'),
                "location": f"{lat},{lon}",
                "temperature": forecast['main']['temp'],
                "humidity": forecast['main']['humidity'],
                "wind_speed": forecast['wind']['speed']
            }
            forecast_data.append(forecast_info)
        
        return forecast_data
    else:
        print("Erreur lors de la récupération des données:", response.status_code)
        return None

def save_to_csv(data_list, filename='weather_forecast_3h.csv'):
    # Vérification si le fichier CSV existe
    header = ["date", "location", "temperature", "humidity", "wind_speed"]
    
    try:
        # Ouverture en mode 'append' pour ajouter les nouvelles données
        with open(filename, mode='a', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=header)
            
            # Si le fichier est vide, on écrit l'en-tête
            file_is_empty = file.tell() == 0
            if file_is_empty:
                writer.writeheader()
                
            # Écriture de chaque ligne de données
            for data in data_list:
                writer.writerow(data)
                
            print(f"Données de prévision sauvegardées dans {filename}")
    except Exception as e:
        print("Erreur lors de la sauvegarde du fichier CSV:", e)

# Exemple d'utilisation
lat = 48.8566  # Latitude de Paris
lon = 2.3522   # Longitude de Paris
api_key = "36669f134b2c1c983dbba764cc582c16"  # Remplacer par ta clé API

# Obtenir les prévisions 3 heures
forecast_data = get_weather_forecast_3h(lat, lon, api_key)

if forecast_data:
    save_to_csv(forecast_data)
```

```python
import requests
import csv
import datetime
import sys
import pandas as pd

# Remplacez 'chemin/vers/votre/projet/module' par le chemin absolu réel vers le dossier "module" via la commande pwd
module_path = '/Users/eliotsouthon/Desktop/POC/Code/Modules'  
sys.path.append(module_path)
import ma_bibliotheque 

def get_weather_forecast_3h(lat, lon, api_key):
    url = f"https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={api_key}&units=metric"
    
    # Requête vers l'API
    response = requests.get(url)
    
    # Vérification du succès de la requête
    if response.status_code == 200:
        data = response.json()
        
        # Extraire les prévisions toutes les 3 heures pour les 5 jours
        forecast_data = []
        for forecast in data['list']:
            forecast_info = {
                "date": datetime.datetime.fromtimestamp(forecast['dt']).strftime('%Y-%m-%d %H:%M:%S'),
                "location": f"{lat},{lon}",
                "temperature": forecast['main']['temp'],
                "humidity": forecast['main']['humidity'],
                "wind_speed": forecast['wind']['speed']
            }
            forecast_data.append(forecast_info)
        
        return forecast_data
    else:
        print("Erreur lors de la récupération des données:", response.status_code)
        return None

def save_to_csv(data_list, filename='weather_forecast_3h.csv'):
    # Vérification si le fichier CSV existe
    header = ["date", "location", "temperature", "humidity", "wind_speed"]
    
    try:
        # Ouverture en mode 'append' pour ajouter les nouvelles données
        with open(filename, mode='a', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=header)
            
            # Si le fichier est vide, on écrit l'en-tête
            file_is_empty = file.tell() == 0
            if file_is_empty:
                writer.writeheader()
                
            # Écriture de chaque ligne de données
            for data in data_list:
                writer.writerow(data)
                
            print(f"Données de prévision sauvegardées dans {filename}")
    except Exception as e:
        print("Erreur lors de la sauvegarde du fichier CSV:", e)

def get_hourly_interpolated_data(lat, lon, api_key, filename='weather_forecast_3h.csv'):
    # Étape 1 : Récupérer les prévisions 3 heures
    forecast_data = get_weather_forecast_3h(lat, lon, api_key)
    
    if forecast_data:
        # Étape 2 : Sauvegarder dans le fichier CSV
        save_to_csv(forecast_data, filename=filename)
        
        # Étape 3 : Charger les données dans un DataFrame
        try:
            df = pd.read_csv(filename)
            
            # Assurer que la colonne "date" est de type datetime
            df['date'] = pd.to_datetime(df['date'])
            
            # Utiliser la fonction d'interpolation de votre bibliothèque
            df_interpolated = ma_bibliotheque.interpolate_to_hourly(df)
            
            return df_interpolated
        except Exception as e:
            print("Erreur lors de la lecture ou de l'interpolation des données:", e)
            return None
    else:
        print("Aucune donnée de prévision récupérée.")
        return None

# Exemple d'utilisation
lat = 48.8566  # Latitude de Paris
lon = 2.3522   # Longitude de Paris
api_key = "36669f134b2c1c983dbba764cc582c16"  # Remplacer par ta clé API

# Obtenir les données interpolées
df_hourly = get_hourly_interpolated_data(lat, lon, api_key)

if df_hourly is not None:
    print("Données interpolées (premières lignes) :")
    print(df_hourly.head())
```
